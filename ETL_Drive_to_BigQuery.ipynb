```python
import pandas as pd
import os
```

## Import Part1 Data

```python
FOLDER_PATH_IN_DRIVE = '/content/drive/MyDrive/Incident Report Data/'

file_list = os.listdir(FOLDER_PATH_IN_DRIVE)

file_list
```

```python
df_files = pd.DataFrame(file_list)

part1_files = df_files[df_files[0].str.contains('(?i)part1\\s*\\.xlsx')]
part2_files = df_files[df_files[0].str.contains('(?i)part2\\s*\\.xlsx')]


print(part1_files)
print(part2_files)
```

```python

# file_name = "Incident Report Dec 2024_Part1.xlsx"
# pd.read_excel(f"{FOLDER_PATH_IN_DRIVE}/{file_name}")
```

```python
list_of_dfs = []

for file in part1_files[0]:

    file_name = file
    df = pd.read_excel(f"{FOLDER_PATH_IN_DRIVE}/{file_name}")
    print(f"read {file_name}")

    list_of_dfs.append(df)



df_part1 = pd.concat(list_of_dfs)

print(len(df_part1))
df_part1.head()
```

```python
df_part1.columns
```

```python
df_part1_cleaned = (

  df_part1.copy()[['1 Number',
                   '2 Company',
                   '3 Priority',
                   '4 Location',
                   '6 Opened',
                   '7 State',
                   'Resolved',
                   '8 Assignment group',
                   '9 Has breached',
                   '10 Category',
                   '11. Subcategory',
                   'Template',
                   'Parent Incident']]


  .rename(columns = {'1 Number' : 'number',
                     '2 Company' : 'company',
                     '3 Priority' : 'priority',
                     '4 Location' : 'location',
                     '6 Opened' : 'open_date',
                     '7 State' : 'state',
                     'Resolved' : 'resolved_date',
                     '8 Assignment group' : 'assignment_group',
                     '9 Has breached' : 'has_breached',
                     '10 Category' : 'category',
                     '11. Subcategory' : 'subcategory',
                     'Template' : 'template',
                     'Parent Incident' : 'parent_incident'})

  .assign(open_date = lambda x: x['open_date'].dt.date,
          resolved_date = lambda x: x['resolved_date'].dt.date)

  .assign(has_breached = lambda x: x['has_breached'].astype('bool'))


)

df_part1_cleaned.head()
```

## Import Part2 Data

```python
list_of_dfs = []

for file in part2_files[0]:

    file_name = file
    df = pd.read_excel(f"{FOLDER_PATH_IN_DRIVE}/{file_name}")
    print(f"read {file_name}")

    list_of_dfs.append(df)



df_part2 = pd.concat(list_of_dfs)

print(len(df_part2))
df_part2.head()
```

```python
df_part2.columns
```

```python
df_part2_cleaned = (

  df_part2.copy()[['Request ID',
                   'Brands',
                   'SLA Name',
                   'Requester',
                   'Created Time',
                   'C.Time',
                   'Request Status',
                   'Resolved Time',
                   'Overdue Status',
                   'Category',
                   'Subcategory']]


  .rename(columns = {'Request ID' : 'number',
                     'Brands' : 'company',
                     'SLA Name' : 'priority',
                     'Requester' : 'location',
                    #  '6 Opened' : 'open_date',
                     'Request Status' : 'state',
                     'Resolved Time' : 'resolved_date',
                     'Overdue Status' : 'has_breached',
                     'Category' : 'category',
                     'Subcategory' : 'subcategory'})

  .assign(number = lambda x: 'INC' + x['number'].astype('str'))

  .assign(open_date = lambda x: pd.to_datetime(x['Created Time']).dt.date)
  .assign(resolved_date = lambda x: pd.to_datetime(x['resolved_date'], format='%d/%m/%Y %H:%M', errors='coerce').dt.date)

  .assign(has_breached = lambda x: x['has_breached'].astype('bool'))

  .drop(columns = ['Created Time','C.Time'])



)

df_part2_cleaned.head()
```

## Combine Part1 + Part2

```python
df_final = pd.concat([df_part1_cleaned, df_part2_cleaned], ignore_index = True)

print(len(df_final))
df_final.head()
```

## Write data to BigQuery

```python
PROJECT_ID = 'lateral-command-464808-e9'
DATASET_ID = 'incident_report' # Replace with your BigQuery Dataset ID
TABLE_ID = 'all_incident_report_data' # Replace with your desired Table Name

destination_table = f"{DATASET_ID}.{TABLE_ID}"


try:
    df_final.to_gbq(
        destination_table = destination_table,
        project_id = PROJECT_ID,
        if_exists = 'replace' # Options: 'fail', 'replace', 'append'
    )
    print(f"\n✅ Successfully uploaded DataFrame to BigQuery table: {TABLE_ID}")

except Exception as e:
    print(f"\n❌ An error occurred during upload: {e}")
```
